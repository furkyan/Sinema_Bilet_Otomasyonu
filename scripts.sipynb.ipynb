# [dbo].[vw_aktif_seanslar]

```sql
/****** Object:  View [dbo].[vw_aktif_seanslar]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Görünümler (views)
CREATE VIEW [dbo].[vw_aktif_seanslar] AS
SELECT 
    s.id as seans_id,
    f.ad as film_adi,
    sl.ad as salon_adi,
    s.tarih,
    s.saat,
    s.fiyat,
    sl.kapasite,
    COUNT(b.id) as satilan_bilet,
    (sl.kapasite - COUNT(b.id)) as bos_koltuk
FROM seanslar s
    INNER JOIN filmler f ON s.film_id = f.id
    INNER JOIN salonlar sl ON s.salon_id = sl.id
    LEFT JOIN biletler b ON s.id = b.seans_id AND b.durum = 'Aktif'
WHERE s.tarih >= CAST(GETDATE() AS DATE)
GROUP BY s.id, f.ad, sl.ad, s.tarih, s.saat, s.fiyat, sl.kapasite;
GO
```

# [dbo].[v_gunluk_satis]

```sql
/****** Object:  View [dbo].[v_gunluk_satis]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[v_gunluk_satis] AS
SELECT 
    CAST(b.satis_tarihi AS DATE) as tarih,
    COUNT(*) as bilet_sayisi,
    SUM(b.fiyat) as toplam_satis,
    AVG(b.fiyat) as ortalama_fiyat
FROM biletler b
WHERE b.durum = 'Aktif'
GROUP BY CAST(b.satis_tarihi AS DATE);
GO
```

# [dbo].[v_aylik_satis]

```sql
/****** Object:  View [dbo].[v_aylik_satis]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[v_aylik_satis] AS
SELECT 
    YEAR(b.satis_tarihi) as yil,
    MONTH(b.satis_tarihi) as ay,
    COUNT(*) as bilet_sayisi,
    SUM(b.fiyat) as toplam_satis,
    AVG(b.fiyat) as ortalama_fiyat
FROM biletler b
WHERE b.durum = 'Aktif'
GROUP BY YEAR(b.satis_tarihi), MONTH(b.satis_tarihi);
GO
```

# [dbo].[v_film_populerlik]

```sql
/****** Object:  View [dbo].[v_film_populerlik]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[v_film_populerlik] AS
SELECT 
    f.id,
    f.ad as film_adi,
    f.tur,
    COUNT(b.id) as bilet_sayisi,
    SUM(b.fiyat) as toplam_hasılat,
    COUNT(DISTINCT s.id) as seans_sayisi
FROM filmler f
LEFT JOIN seanslar s ON f.id = s.film_id
LEFT JOIN biletler b ON s.id = b.seans_id AND b.durum = 'Aktif'
GROUP BY f.id, f.ad, f.tur;
GO
```

# [dbo].[v_salon_doluluk]

```sql
/****** Object:  View [dbo].[v_salon_doluluk]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[v_salon_doluluk] AS
SELECT 
    sl.id,
    sl.ad as salon_adi,
    sl.kapasite,
    COUNT(s.id) as toplam_seans,
    COUNT(b.id) as satilan_bilet,
    CASE 
        WHEN COUNT(s.id) > 0 THEN 
            CAST(COUNT(b.id) AS FLOAT) / (COUNT(s.id) * sl.kapasite) * 100
        ELSE 0 
    END as doluluk_orani
FROM salonlar sl
LEFT JOIN seanslar s ON sl.id = s.salon_id
LEFT JOIN biletler b ON s.id = b.seans_id AND b.durum = 'Aktif'
GROUP BY sl.id, sl.ad, sl.kapasite;
GO
```

# [dbo].[fn_ozet_istatistik]

```sql
/****** Object:  UserDefinedFunction [dbo].[fn_ozet_istatistik]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE FUNCTION [dbo].[fn_ozet_istatistik]()
RETURNS TABLE
AS
RETURN
(
    SELECT 
        (SELECT COUNT(*) FROM filmler) as toplam_film,
        (SELECT COUNT(*) FROM salonlar) as toplam_salon,
        (SELECT COUNT(*) FROM kullanicilar WHERE rol = 'Müşteri') as toplam_musteri,
        (SELECT COUNT(*) FROM seanslar WHERE tarih >= CAST(GETDATE() AS DATE)) as aktif_seans,
        (SELECT ISNULL(SUM(fiyat), 0) FROM biletler WHERE CAST(satis_tarihi AS DATE) = CAST(GETDATE() AS DATE) AND durum = 'Aktif') as gunluk_satis,
        (SELECT ISNULL(SUM(fiyat), 0) FROM biletler WHERE MONTH(satis_tarihi) = MONTH(GETDATE()) AND YEAR(satis_tarihi) = YEAR(GETDATE()) AND durum = 'Aktif') as aylik_satis
);
GO
```

# [dbo].[v_populer_filmler]

```sql
/****** Object:  View [dbo].[v_populer_filmler]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[v_populer_filmler] AS
SELECT TOP 10
    f.ad as film_adi,
    f.tur,
    COUNT(b.id) as izlenme_sayisi,
    SUM(b.fiyat) as hasılat,
    AVG(b.fiyat) as ortalama_bilet_fiyati
FROM filmler f
INNER JOIN seanslar s ON f.id = s.film_id
INNER JOIN biletler b ON s.id = b.seans_id
WHERE b.durum = 'Aktif'
GROUP BY f.id, f.ad, f.tur
ORDER BY COUNT(b.id) DESC;
GO
```

# [dbo].[v_haftalik_trend]

```sql
/****** Object:  View [dbo].[v_haftalik_trend]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE VIEW [dbo].[v_haftalik_trend] AS
SELECT 
    DATEPART(WEEK, b.satis_tarihi) as hafta_no,
    YEAR(b.satis_tarihi) as yil,
    COUNT(*) as bilet_sayisi,
    SUM(b.fiyat) as toplam_satis,
    COUNT(DISTINCT CAST(b.satis_tarihi AS DATE)) as aktif_gun_sayisi
FROM biletler b
WHERE b.durum = 'Aktif'
    AND b.satis_tarihi >= DATEADD(WEEK, -12, GETDATE()) -- Son 12 hafta
GROUP BY DATEPART(WEEK, b.satis_tarihi), YEAR(b.satis_tarihi);
GO
```

# [dbo].[CreateSeatsForSalon]

```sql
/****** Object:  StoredProcedure [dbo].[CreateSeatsForSalon]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Koltukları oluşturmak için stored procedure
CREATE PROCEDURE [dbo].[CreateSeatsForSalon]
    @salon_id INT,
    @satir INT,
    @sutun INT,
    @vip BIT
AS
BEGIN
    DECLARE @i INT = 1;
    DECLARE @j INT = 1;
    DECLARE @tip NVARCHAR(20);
    
    WHILE @i <= @satir
    BEGIN
        SET @j = 1;
        WHILE @j <= @sutun
        BEGIN
            -- Koltuk tipini belirle
            IF @vip = 1
                SET @tip = 'VIP';
            ELSE IF @i = @satir
                SET @tip = 'CIFT';
            ELSE
                SET @tip = 'NORMAL';
                
            INSERT INTO koltuklar (salon_id, sira_no, koltuk_no, tip)
            VALUES (@salon_id, @i, @j, @tip);
            
            SET @j = @j + 1;
        END
        SET @i = @i + 1;
    END
END;
GO
```

# [dbo].[sp_detayli_satis_raporu]

```sql
/****** Object:  StoredProcedure [dbo].[sp_detayli_satis_raporu]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_detayli_satis_raporu]
    @baslangic_tarihi DATE = NULL,
    @bitis_tarihi DATE = NULL
AS
BEGIN
    -- Eğer tarih belirtilmemişse son 30 günü al
    IF @baslangic_tarihi IS NULL
        SET @baslangic_tarihi = DATEADD(DAY, -30, GETDATE());
    
    IF @bitis_tarihi IS NULL
        SET @bitis_tarihi = GETDATE();
    
    SELECT 
        CAST(b.satis_tarihi AS DATE) as tarih,
        f.ad as film_adi,
        sl.ad as salon_adi,
        b.koltuk_no,
        b.fiyat,
        b.durum,
        k.ad + ' ' + k.soyad as musteri_adi
    FROM biletler b
    INNER JOIN seanslar s ON b.seans_id = s.id
    INNER JOIN filmler f ON s.film_id = f.id
    INNER JOIN salonlar sl ON s.salon_id = sl.id
    INNER JOIN kullanicilar k ON b.kullanici_id = k.id
    WHERE CAST(b.satis_tarihi AS DATE) BETWEEN @baslangic_tarihi AND @bitis_tarihi
    ORDER BY b.satis_tarihi DESC;
END;
GO
```

# [dbo].[sp_GetDailySales]

```sql
/****** Object:  StoredProcedure [dbo].[sp_GetDailySales]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_GetDailySales]
    @tarih DATE = NULL
AS
BEGIN
    IF @tarih IS NULL
        SET @tarih = CAST(GETDATE() AS DATE);
        
    SELECT 
        COUNT(*) as toplam_bilet,
        SUM(fiyat) as toplam_gelir,
        AVG(fiyat) as ortalama_fiyat
    FROM biletler 
    WHERE CAST(satis_tarihi AS DATE) = @tarih 
        AND durum = 'Aktif';
END;
GO
```

# [dbo].[sp_GetUserTickets]

```sql
/****** Object:  StoredProcedure [dbo].[sp_GetUserTickets]    Script Date: 24.05.2025 00:09:52 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Stored procedure'ler
CREATE PROCEDURE [dbo].[sp_GetUserTickets]
    @kullanici_id INT
AS
BEGIN
    SELECT 
        b.id as bilet_id,
        f.ad as film_adi,
        sl.ad as salon_adi,
        s.tarih,
        s.saat,
        b.koltuk_no,
        b.fiyat,
        b.satis_tarihi,
        b.durum
    FROM biletler b
        INNER JOIN seanslar s ON b.seans_id = s.id
        INNER JOIN filmler f ON s.film_id = f.id
        INNER JOIN salonlar sl ON s.salon_id = sl.id
    WHERE b.kullanici_id = @kullanici_id
    ORDER BY b.satis_tarihi DESC;
END;
GO
```
